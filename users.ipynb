```python
from pathlib import Path
import json
from collections import defaultdict

def get_value_counts(root_folder, key):
    root = Path(root_folder)
    value_counts = defaultdict(int)

    for file in root.rglob("*.jsonl"):
        with file.open("r", encoding="utf-8") as f:
            for line in f:
                try:
                    data = json.loads(line)
                    val = data.get(key)
                    if val is not None:
                        value_counts[val] += 1
                except json.JSONDecodeError:
                    continue

    # sort by count (desc), then by key (asc) for stability
    return sorted(value_counts.items(), key=lambda x: (-x[1], x[0]))
```

```python
sorted_value_counts = get_value_counts("/Users/vishalsankarram/dsci558-project/game_review_batches", "username")
with open("users.txt", "w") as f:
    for k,v in sorted_value_counts:
        f.write(str(k) + "\n")
with open("users_count.txt", "w") as f:
    for k,v in sorted_value_counts:
        f.write(str(k) + "\t"+ str(v) + "\n")
print(len(sorted_value_counts))
```

```python
from pathlib import Path
import json
from collections import defaultdict

def get_count_and_game_counts(root_folder, key):
    root = Path(root_folder)
    
    total_count = 0
    game_counts = defaultdict(int)  # bgg_id -> count

    for file in root.rglob("*.jsonl"):
        with open(file, "r", encoding="utf-8") as f:
            for line in f:
                try:
                    data = json.loads(line)
                    val = data.get(key)

                    if val == "1" and key != "comment" and key != "user_rating":
                        total_count += 1
                        game_id = data.get("bgg_id")
                        if game_id:
                            game_counts[game_id] += 1
                    elif key == "comment" and data.get("comment") != "":
                        total_count += 1
                        game_id = data.get("bgg_id")
                        if game_id:
                            game_counts[game_id] += 1
                    elif key == "user_rating" and data.get("user_rating") is not None:
                        total_count += 1
                        game_id = data.get("bgg_id")
                        if game_id:
                            game_counts[game_id] += 1
                except:
                    continue

    return total_count, dict(game_counts)
```

```python
root = "/Users/vishalsankarram/dsci558-project/user"

keys = [
    "want",
    "wishlist",
    "want_to_play",
    "want_to_buy",
    "own",
    "for_trade",
    "comment",
    "user_rating"
]

for key in keys:
    total, game_counts = get_count_and_game_counts(root, key)
    unique_games = len(game_counts)

    print(f"{key}:")
    print("  Total:", total)
    print("  Unique games:", unique_games)
```
